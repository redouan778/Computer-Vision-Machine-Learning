```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('../data/Life Expectancy Data.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
X = df.drop('Country', axis=1)
X.head()
```

```python
X.isna().sum()
```

```python
for ls in X.columns:
    X[ls].fillna(value=X[ls].mean(),inplace=True)
```

```python
X.isna().sum()
```

```python
target_countries = df['Country'].unique()
len(target_countries)
```

```python
target_dict = {n:i for i, n in enumerate(target_countries)}
target_dict
```

```python
y= df['Country'].map(target_dict)
y.head()
```

```python
from tensorflow.keras.utils import to_categorical
y_cat = to_categorical(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y_cat)
```

```python
input_count = len(df.columns)-1
output_count = len(target_countries)
```

```python
model = Sequential()

model.add(Dense(193, input_shape=(21,), activation='softmax'))


model.compile(Adam(learning_rate=0.2),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=2000, validation_split=0.1)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print("RMSE=",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print("R^2=",metrics.r2_score(y_test,y_pred))
```
